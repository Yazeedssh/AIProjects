### Dealing with NAs

```python
def count_NAs(df, perc):
    NaList = []
    numOfRows = len(df)
    for i in df.columns:
        numOfNAs = df[i].isna().sum()
        print(f'{i:<30}, Rows: {numOfRows:>5}, NAs: {numOfNAs:<25}, precentage: {round(numOfNAs/numOfRows * 100.00,2):<25} , Data Type : {df[i].dtype}')
        if ((numOfNAs/numOfRows) * 100.000) >= perc:
           NaList.append(i)
    print(f'columns with NAs over {perc}%: {NaList}')
    return NaList
# Display NAs
cutoff = 40
NAs = count_NAs(df, cutoff)
print(f"Total with NAs over {cutoff}% column number :",len(NAs))
```

```python
def fill_cata_nulls(df , col_name) :
  mode_value = df[col_name].mode()[0]
  return df[col_name].fillna(mode_value)
```

```python
df['Program Sub Category Code'] = fill_cata_nulls(df, "Program Sub Category Code")
df["Program Skill Level"] = fill_cata_nulls(df , "Program Skill Level")
df["Education Speaciality"] = fill_cata_nulls(df , "Education Speaciality")
df["Level of Education"] = fill_cata_nulls(df , "Level of Education")
df["Employment Status"] = fill_cata_nulls(df , "Employment Status")
df["Home Region"] = fill_cata_nulls(df , "Home Region")
df["Home City"] = fill_cata_nulls(df , "Home City")
df['Y'] = fill_cata_nulls(df , "Y")
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
# labels = np.ravel(labels)
# labels
def train_test() :
  X = df
  y = labels

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
  smote = SMOTE(random_state=42)
  X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
  return X_train_resampled, X_test, y_train_resampled, y_test
```

## Evaluation Functions

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True))
        print("Train Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("\n_______________________________________________")


    elif train==False:
        print("Test Result:\n================================================")
        pred = clf.predict(X_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))

        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
import seaborn as sns
import matplotlib.pyplot as plt
def cross_accuracy(model ,X_train , y_train) :
  print("Test Result:\n================================================")
  print("\n================================================")
  accuracy_scores = cross_val_score(model, X_train, y_train, cv=3, scoring="accuracy")
  print("Cross-Validation Accuracy Scores:", accuracy_scores)

  # cross_val_predict to get cross-validated predictions
  y_train_pred = cross_val_predict(model, X_train, y_train, cv=3)

  # Generate confusion matrix
  conf_matrix = confusion_matrix(y_train, y_train_pred)
  class_names = labels
  df_cm = pd.DataFrame(conf_matrix, index=class_names, columns=class_names)
  plt.figure(figsize=(8, 6))
  plt.title('Confusion Matrix')
  sns.heatmap(df_cm, annot=True, cmap='Blues', fmt='g')
  plt.xlabel('Predicted')
  plt.ylabel('Actual')
  plt.show()
```

## Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

X_train, X_test, y_train, y_test = train_test()
# scalar = StandardScaler()
log_reg = LogisticRegression(random_state=101 , class_weight='balanced')
scalar = StandardScaler()

operations = [("scale", scalar), ("logistic", log_reg)]

pipe_model = Pipeline(steps=operations)


# Step 4: Train Pipeline
pipe_model.fit(X_train, y_train)
```

```python
y_pred = pipe_model.predict(X_test)
print(y_pred[100:500])
```

```python
print_score(pipe_model, X_train, y_train, X_test, y_test, train=True)
print_score(pipe_model, X_train, y_train, X_test, y_test, train=False)
```

```python
logistic_probs = pipe_model.predict_proba(X_test)[:, 1]
logistic_probs
```

## K-Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test()
knn = KNeighborsClassifier(n_neighbors=5 , weights ='distance')
knn.fit(X_train , y_train)
```

## SVM

```python
from sklearn.svm import LinearSVC
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test()
poly_kernel_svm_clf = Pipeline([
    # Step 1: Feature scaling using StandardScaler
    ("scaler", StandardScaler()),

    # Step 2: SVM classifier with Polynomial Kernel
    ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5 , class_weight='balanced' , probability=True))
])


# Fitting the classifier to the data
poly_kernel_svm_clf.fit(X_train, y_train)
```

```python
print_score(poly_kernel_svm_clf, X_train, y_train, X_test, y_test, train=True)
print_score(poly_kernel_svm_clf, X_train, y_train, X_test, y_test, train=False)
```

```python
cross_accuracy(poly_kernel_svm_clf , X_train , y_train)
```

```python
svm_probs = poly_kernel_svm_clf.predict_proba(X_test)[:, 1]
svm_probs
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

X_train, X_test, y_train, y_test = train_test()

knn_fpr, knn_tpr, _ = roc_curve(y_test, knn_probs)
knn_auc = auc(knn_fpr, knn_tpr)

# Logistic Regression
logistic_fpr, logistic_tpr, _ = roc_curve(y_test, logistic_probs)
logistic_auc = auc(logistic_fpr, logistic_tpr)

# SVM
svm_fpr, svm_tpr, _ = roc_curve(y_test, svm_probs)
svm_auc = auc(svm_fpr, svm_tpr)

# 4. Plot ROC curves for all models
plt.figure(figsize=(8, 6))
plt.plot(knn_fpr, knn_tpr, color='blue', lw=2, label=f'KNN (AUC = {knn_auc:.2f})')
plt.plot(logistic_fpr, logistic_tpr, color='green', lw=2, label=f'Logistic Regression (AUC = {logistic_auc:.2f})')
plt.plot(svm_fpr, svm_tpr, color='red', lw=2, label=f'SVM (AUC = {svm_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```
